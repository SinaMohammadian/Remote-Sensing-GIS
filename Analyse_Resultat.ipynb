```python
dataframe = pd.read_excel(r"C:\Users\LENOVO\OneDrive\Desktop\Projet_SIG\choix_altimeter.xlsx")
```

```python
dataframe
```

```python
dataframe["time"] = pd.to_datetime(dataframe["time"])

plt.figure(figsize=(10, 5))
plt.plot(dataframe["time"], dataframe["adt"])
plt.xlabel("Temps", fontsize=17)
plt.ylabel("ADT (m)", fontsize=17)
plt.title("Variation de l'ADT (Absolute Dynamic Topography) en fonction du temps", fontsize=17)
# plt.grid(True)

sample_indices = [0, len(dataframe)//3, 2*len(dataframe)//3, len(dataframe)-1]
plt.xticks([dataframe["time"].iloc[i] for i in sample_indices],
           [dataframe["time"].iloc[i].strftime("%Y-%m-%d") for i in sample_indices], fontsize=13)
plt.yticks(fontsize=13)
plt.tight_layout()
plt.show()
```

# Diagramme de comparaison des zones

```python
labels = ['Très faible', 'Faible', 'Modérée', 'Élevée', 'Très élevée']
areas = [275, 399, 398, 180, 53]
colors = ['#1a9641',  '#a6d96a',  '#ffffc0',  '#fdae61',  '#d7191c']

total = sum(areas)
percentages = [100 * a / total for a in areas]

plt.figure(figsize=(8, 6))
bars = plt.bar(labels, areas, color=colors)
plt.xlabel("Niveau de la vulnérabilité", fontsize=18)
plt.ylabel("Surface (km²)", fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title("Surface des zones par niveau de vulnérabilité côtière", fontsize=18)
plt.grid(axis='y')

for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height / 2,  
             f"{percentages[i]:.1f}%",
             ha='center', va='center',
             fontsize=22, color='black', fontweight='bold')

plt.tight_layout()
plt.show()
```
